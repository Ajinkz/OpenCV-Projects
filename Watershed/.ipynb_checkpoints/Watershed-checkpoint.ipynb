# Watershed

Bir masanın üstünde bozuk paralar olduğunu düşünelim. Bozuk paralar arasında yeterli bir mesafe varsa kontör metodu ile paraları bulmakta sıkıntı yaşamayız. Fakat paralar arasında ki mesafe çok az veya paralar birbirine değiyorsa bu durumda kontör metodu ile bulma yöntemi düzgün çalışmayacaktır.

```python
# Gerekli kütüphanelerin yüklenmesi
import matplotlib.pyplot as plt
import numpy as np
import cv2

"""
Fotoğrafın yüklenmesi ve pyrMeanShiftFiltering uygulanması.

pyrMeanShiftFiltering
-> Bir görüntünün ortalama kaydırma segmentasyonunun ilk adımını gerçekleştirir.
-> Parametreler:
    - src: görüntü.
    - sp: uzaysal pencere yarıçapı. Varsayılan 5.
    - sr: renk penceresinin yarıçapı. Varsayılan 10.
    
"""
image = cv2.imread("./images/turk_prices.jpeg")
shifted = cv2.pyrMeanShiftFiltering(image, 21, 51)
plt.figure(figsize=(12,8))
plt.imshow(shifted)
```

```python
# Burada threshold kısmında THRESH_OTSU kullanmamızın nedeni yukarıda paraları gri
# tonlamaya çevirdiğimizden kaynaklanıyor.
gray = cv2.cvtColor(shifted, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
plt.figure(figsize=(12,8))
plt.imshow(thresh)
```

```python
# En son adımda ise klasik yöntemimiz olan kontörleri bulmaya geldik
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[1]
print("Toplam bulunan para sayısı: {}".format(len(cnts)))
```

```python
# 3 göstermesine rağmen
for (i, c) in enumerate(cnts):
	((x, y), _) = cv2.minEnclosingCircle(c)
	cv2.putText(image, "#{}".format(i + 1), (int(x) - 10, int(y)),
		cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 6)
	cv2.drawContours(image, [c], -1, (0, 255, 0), 4)
    
plt.figure(figsize=(12,8))
plt.imshow(image)
```

Şimdi gelelim Watershed uygulamaya. 

```python
# gerekli kütüphanaler
from scipy import ndimage
from skimage.morphology import watershed
from skimage.feature import peak_local_max
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# fotoğrafın yüklenmesi ve pyrMeanShiftFiltering uygulanması
image = cv2.imread("./images/turk_prices.jpeg")
shifted = cv2.pyrMeanShiftFiltering(image, 21, 51)

# gri tonlamaya dönüştürme ve eşikleme
gray = cv2.cvtColor(shifted, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

"""
peak_local_max
-> Peak_local_max işlevi, bir görüntüdeki yerel zirvelerin (maksimum) koordinatlarını döndürür. 
-> Yerel maksimumları bulmak için maksimum filtre kullanılır.
-> Bu işlem, orijinal görüntüyü genişletir ve komşu yerel maksimumları genişleme boyutundan daha yakın birleştirir.
-> Orijinal görüntünün genişletilmiş görüntüye eşit olduğu konumlar yerel maksimumlar olarak döndürülür.
"""
D = ndimage.distance_transform_edt(thresh)
localMax = peak_local_max(D, indices=False, min_distance=30, labels=thresh)

markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-D, markers, mask=thresh)
print("Toplam bulunan para sayısı: {}".format(len(np.unique(labels)) - 1))

fig, axes = plt.subplots(ncols=2 ,figsize=(15,8))
ax0, ax1 = axes
ax0.imshow(-D,  cmap=plt.cm.jet, interpolation='nearest')
ax1.imshow(labels,  cmap=plt.cm.jet, interpolation='nearest')
for ax in axes:
    ax.axis('off')
plt.show()
```

```python
for label in np.unique(labels):
	if label == 0:
		continue
	mask = np.zeros(gray.shape, dtype="uint8")
	mask[labels == label] = 255
	cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[1]
	c = max(cnts, key=cv2.contourArea)
	((x, y), r) = cv2.minEnclosingCircle(c)
	cv2.circle(image, (int(x), int(y)), int(r), (0, 255, 0), 6)
	cv2.putText(image, "#{}".format(label), (int(x) - 10, int(y)),
		cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 4)
    
plt.figure(figsize=(12,8))
plt.imshow(image)
```
